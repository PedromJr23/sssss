```python
# passo a passo 
# Passo 1: Importar a base de dados
import pandas as pd

# Passo 2: Visualizar a base 
tabela = pandas.read_csv("cancelamentos_sample.csv")
tabela = tabela.drop(columns="CustomerID")
display(tabela)
```

```python
# identificando e removendo valores vazios
tabela =tabela.dropna()
display(tabela.info())
```

```python
# Passo 4: Análisar a causa do cancelamento (% de cancelamentos)
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
# analisando o contrato mensal
agrupamento = tabela.groupby("duracao_contrato").mean(numeric_only=True)
display(agrupamento)
# descobrimos aqui que a média de cancelamentos é 1, ou seja, praticamente todos os contratos mensais cancelaram (ou todos)
```

```python
display(tabela["tempo_como_cliente"].value_counts(normalize=True).map("{:.1%}".format))

display(tabela["frequencia_uso"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
display(tabela["duracao_contrato"].value_counts(normalize=True).map("{:.1%}".format))


display(tabela["total_gasto"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
tabela = tabela[tabela["duracao_contrato"]!="Monthly"]
display(tabela)
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
# chegamos agora em menos da metade de pessoas cancelando, mas ainda temos muitas pessoas ai, vamos continuar analisando
display(tabela["assinatura"].value_counts(normalize=True))
display(tabela.groupby("assinatura").mean(numeric_only=True))
# vemos que assinatura é quase 1/3, 1/3, 1/3
# e que os cancelamentos são na média bem parecidos, então fica difícil tirar alguma conclusão da média, vamos precisar ir mais a fundo
```

```python
# vamos criar gráfico, porque só com números tá difícil de visualizar
import plotly.express as px

for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color="cancelou")
    grafico.show()
```

```python
# ligações para call center acima de 4, cancelam?
# dias de atraso > 2, todos cancelam !

tabela = tabela[tabela["ligacoes_callcenter"] < 5]
tabela = tabela[tabela["dias_atraso"] <= 20]

display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```
